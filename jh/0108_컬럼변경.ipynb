# 데이터가 5000개가 넘어가서 샤리포 기법 효율이 떨어짐 K-S검증으로 전환

```python
# 필요한 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kstest

# 가입 기간 정규성 검정
stat, p = kstest(tmp1['가입 기간'].dropna(), 'norm', args=(tmp1['가입 기간'].mean(), tmp1['가입 기간'].std()))
print(f"Kolmogorov-Smirnov 검정 (가입 기간): 통계량={stat:.3f}, p-값={p:.3f}")
if p > 0.05:
    print("가입 기간 데이터는 정규성을 띕니다.")
else:
    print("가입 기간 데이터는 정규성을 띄지 않습니다.")

# 주문 횟수 정규성 검정
stat, p = kstest(tmp1['주문 횟수'].dropna(), 'norm', args=(tmp1['주문 횟수'].mean(), tmp1['주문 횟수'].std()))
print(f"Kolmogorov-Smirnov 검정 (주문 횟수): 통계량={stat:.3f}, p-값={p:.3f}")
if p > 0.05:
    print("주문 횟수 데이터는 정규성을 띕니다.")
else:
    print("주문 횟수 데이터는 정규성을 띄지 않습니다.")

# 가입 기간 히스토그램 및 Q-Q 플롯
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(tmp1['가입 기간'].dropna(), kde=True)
plt.title('가입 기간 히스토그램')

plt.subplot(1, 2, 2)
sns.scatterplot(x=np.sort(tmp1['가입 기간'].dropna()), y=np.sort(tmp1['가입 기간'].dropna()))
plt.title('가입 기간 Q-Q 플롯')

plt.tight_layout()
plt.show()

# 주문 횟수 히스토그램 및 Q-Q 플롯
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(tmp1['주문 횟수'].dropna(), kde=True)
plt.title('주문 횟수 히스토그램')

plt.subplot(1, 2, 2)
sns.scatterplot(x=np.sort(tmp1['주문 횟수'].dropna()), y=np.sort(tmp1['주문 횟수'].dropna()))
plt.title('주문 횟수 Q-Q 플롯')

plt.tight_layout()
plt.show()
```

# 추가로 A-D 검정결과

```python
import matplotlib.pyplot as plt
import numpy as np  # NumPy 임포트 추가

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 필요한 라이브러리 임포트
from scipy.stats import anderson
import seaborn as sns

# 가입 기간 A-D 검정
result = anderson(tmp1['가입 기간'].dropna(), dist='norm')
print(f"Anderson-Darling 검정 (가입 기간): 통계량={result.statistic:.3f}")
print("임계값(Critical Values):", result.critical_values)
print("유의수준(Significance Levels):", result.significance_level)

# A-D 검정 결과 해석
if result.statistic < result.critical_values[2]:  # 유의수준 5% 기준
    print("가입 기간 데이터는 정규성을 띕니다.")
else:
    print("가입 기간 데이터는 정규성을 띄지 않습니다.")

# 가입 기간 히스토그램 및 Q-Q 플롯
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(tmp1['가입 기간'].dropna(), kde=True)
plt.title('가입 기간 히스토그램')

plt.subplot(1, 2, 2)
sns.scatterplot(x=np.sort(tmp1['가입 기간'].dropna()), y=np.sort(tmp1['가입 기간'].dropna()))
plt.plot(np.sort(tmp1['가입 기간'].dropna()), np.sort(tmp1['가입 기간'].dropna()), color='red', linestyle='--')  # 빨간 정규분위선 추가
plt.title('가입 기간 Q-Q 플롯')

plt.tight_layout()
plt.show()

# 주문 횟수 A-D 검정
result = anderson(tmp1['주문 횟수'].dropna(), dist='norm')
print(f"Anderson-Darling 검정 (주문 횟수): 통계량={result.statistic:.3f}")
print("임계값(Critical Values):", result.critical_values)
print("유의수준(Significance Levels):", result.significance_level)

# A-D 검정 결과 해석
if result.statistic < result.critical_values[2]:  # 유의수준 5% 기준
    print("주문 횟수 데이터는 정규성을 띕니다.")
else:
    print("주문 횟수 데이터는 정규성을 띄지 않습니다.")

# 주문 횟수 히스토그램 및 Q-Q 플롯
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(tmp1['주문 횟수'].dropna(), kde=True)
plt.title('주문 횟수 히스토그램')

plt.subplot(1, 2, 2)
sns.scatterplot(x=np.sort(tmp1['주문 횟수'].dropna()), y=np.sort(tmp1['주문 횟수'].dropna()))
plt.plot(np.sort(tmp1['주문 횟수'].dropna()), np.sort(tmp1['주문 횟수'].dropna()), color='red', linestyle='--')  # 빨간 정규분위선 추가
plt.title('주문 횟수 Q-Q 플롯')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import seaborn as sns

# 스피어만 상관계수 계산
correlation, p_value = spearmanr(tmp1['주문 횟수'].dropna(), tmp1['가입 기간'].dropna())

# 결과 출력
print(f"스피어만 상관계수: {correlation:.3f}, p-value: {p_value:.3f}")

# 상관관계 해석
if p_value < 0.05:
    print("주문 횟수와 가입 기간 간에 유의미한 상관관계가 있습니다.")
else:
    print("주문 횟수와 가입 기간 간에 유의미한 상관관계가 없습니다.")

# 시각화 추가
plt.figure(figsize=(10, 6))
sns.scatterplot(x=tmp1['주문 횟수'].dropna(), y=tmp1['가입 기간'].dropna())
plt.title('주문 횟수와 가입 기간 간의 산점도')
plt.xlabel('주문 횟수')
plt.ylabel('가입 기간')
plt.axhline(y=tmp1['가입 기간'].mean(), color='red', linestyle='--', label='가입 기간 평균')
plt.axvline(x=tmp1['주문 횟수'].mean(), color='blue', linestyle='--', label='주문 횟수 평균')
plt.legend()

# 코렐레이션 시각화 추가
plt.figure(figsize=(8, 6))
sns.heatmap(tmp1[['주문 횟수', '가입 기간']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('주문 횟수와 가입 기간 간의 상관관계 히트맵')
plt.show()
```

```python
from scipy.stats import kendalltau
import matplotlib.pyplot as plt
import seaborn as sns

# 켄달의 타우 상관계수 계산
tau, p_value_tau = kendalltau(tmp1['주문 횟수'].dropna(), tmp1['가입 기간'].dropna())

# 결과 출력
print(f"켄달의 타우 상관계수: {tau:.3f}, p-value: {p_value_tau:.3f}")

# 상관관계 해석
if p_value_tau < 0.05:
    print("주문 횟수와 가입 기간 간에 유의미한 상관관계가 있습니다.")
else:
    print("주문 횟수와 가입 기간 간에 유의미한 상관관계가 없습니다.")

# 히트맵 시각화 추가
plt.figure(figsize=(8, 6))
sns.heatmap(tmp1[['주문 횟수', '가입 기간']].corr(method='kendall'), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('주문 횟수와 가입 기간 간의 켄달의 타우 상관관계 히트맵')
plt.show()
```

장점:
데이터의 전반적인 구조 파악 가능:

히트맵은 변수 간의 상관성을 시각적으로 확인할 수 있어, 변수들 간의 연관성을 빠르게 파악할 수 있습니다.
어떤 변수가 서로 강한 상관관계를 가지는지 확인하여, 분석에서 중요한 변수와 덜 중요한 변수를 선별하는 데 도움을 줍니다.
전처리 우선순위 설정:

상관관계가 약하거나 불필요한 변수들을 먼저 제거할지, 또는 상관성이 높은 변수들에 대해 추가 조치를 취할지를 결정하는 데 유용합니다.
이상치 및 결측치 전처리 우선순위 판단:

상관관계가 높은 변수에 대해 우선적으로 결측치나 이상치를 처리하고, 상관성이 낮은 변수는 나중에 처리하거나 제거할 수 있습니다.
2. 유의점 (주의사항)
히트맵으로 상관관계를 확인하기 전에 전처리를 생략하면 다음과 같은 문제가 발생할 수 있습니다:

2.1. 결측치(Missing Values)
결측치가 있는 변수는 상관계수 계산 시 문제가 발생할 수 있습니다.
일반적으로 상관계수를 계산하기 위해서는 결측치를 제거하거나, 임의로 대체해야 합니다.
2.2. 이상치(Outliers)
이상치는 상관계수 계산에 큰 영향을 줄 수 있습니다.
전처리를 생략하면 상관계수가 왜곡될 위험이 있습니다.
2.3. 변수 스케일 차이
변수 간의 스케일(단위)이 크게 다르면 상관관계가 왜곡될 수 있습니다.
예를 들어, 변수 하나는 01 범위이고 다른 변수는 1,00010,000 범위라면, 전처리 없이 상관관계를 확인하는 것은 부정확할 수 있습니다.
2.4. 비선형 관계
상관계수(특히 Pearson)는 선형 관계를 측정합니다.
비선형적인 관계는 상관계수에서 제대로 반영되지 않을 수 있으므로, 비선형 관계를 고려해야 하는 경우 전처리를 통한 변환(예: 로그 변환)이 필요합니다.
3. 추천 접근법
히트맵을 사용하기 전, 다음과 같은 최소한의 전처리를 권장합니다:

3.1. 결측치 처리
결측치를 삭제하거나, 평균/중앙값/최빈값 등으로 대체한 뒤 상관관계를 확인합니다.
예: df.fillna(df.mean(), inplace=True)
3.2. 이상치 처리
이상치를 사전에 탐지하여 제거하거나 상한/하한 값을 지정합니다.
예: IQR(Interquartile Range)을 활용한 이상치 제거.
3.3. 변수 스케일 조정
변수의 스케일 차이가 큰 경우, 정규화(Standardization)나 정규 분포화(Normalization)를 수행한 후 상관관계를 계산합니다.
예: from sklearn.preprocessing import StandardScaler
