```python
import pandas as pd

# 데이터 로드
file_path = '쇼핑몰이탈.csv'  # 파일 경로를 지정하세요
data = pd.read_csv(file_path)

# 데이터 구조 확인
print("데이터셋 구조:")
print(data.info())

# 데이터 샘플 확인
print("\n데이터 샘플:")
print(data.head())

# 데이터의 크기 확인
print("\n데이터 크기 (행, 열):", data.shape)
```

```python
# 기술통계량 확인
print("\n기술통계량:")
print(data.describe())
```

```python
import matplotlib
matplotlib.rc('font', family='Malgun Gothic')

import matplotlib.pyplot as plt

# 회원ID 열 제외하고 히스토그램 시각화
fig, axs = plt.subplots(5, 4, figsize=(20, 15))
for i, column in enumerate(df.columns[1:]):
    axs[i//4, i%4].hist(data[column], bins=50)
    axs[i//4, i%4].set_title(column + ' 히스토그램')
plt.tight_layout()
plt.show()

```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'

# 수치형 열만 선택하여 박스플롯 시각화
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
n_cols = len(numeric_columns)
rows = (n_cols + 3) // 4  # 4개의 열로 구성하되, 필요한 행 수 계산

fig, axs = plt.subplots(rows, 4, figsize=(20, rows * 3))
for i, column in enumerate(numeric_columns):
    row, col = i // 4, i % 4
    axs[row, col].boxplot(data[column])
    axs[row, col].set_title(column + ' 박스플롯')

# 사용하지 않는 서브플롯 제거
for i in range(n_cols, rows * 4):
    row, col = i // 4, i % 4
    fig.delaxes(axs[row, col])

plt.tight_layout()
plt.show()
```

```python
# Z-score 기반 극단값 확인
z_scores = {}
z_outliers = set()
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    if column != '고객 ID':
        z_score = (df[column] - df[column].mean()) / df[column].std()
        z_scores[column] = z_score
        z_outliers.update(df[abs(z_score) > 3].index)

# IQR 기반 극단값 확인  
iqr_outliers = set()
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    if column != '고객 ID':
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        iqr_outliers.update(df[(df[column] < lower_bound) | (df[column] > upper_bound)].index)

# 두 방법으로 검출된 극단값 통합
outliers = z_outliers.union(iqr_outliers)

print(f"Z-score 기반 극단값 수: {len(z_outliers)}")
print(f"IQR 기반 극단값 수: {len(iqr_outliers)}")
print(f"전체 극단값 수: {len(outliers)}")

# 극단값이 있는 행 제거한 데이터프레임 생성
df_cleaned = df[~df.index.isin(outliers)]
print(f"\n원본 데이터 크기: {df.shape}")
print(f"정제된 데이터 크기: {df_cleaned.shape}")

df2 = df_cleaned
df2
```
